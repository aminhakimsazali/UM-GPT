## 1. Extract PDF from SC Website

```python
websites = [
"https://www.sc.com.my/regulation/regulatory-faqs" , 
"https://www.sc.com.my/regulation/technical-notes-practice-notes-and-circulars",
"https://www.sc.com.my/regulation/guidance-notes-and-guiding-principles",
"https://www.sc.com.my/regulation/acts"    
]
```

```python
from tqdm import tqdm
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin

def is_valid_url(url):
    try:
        result = urlparse(url)
        return all([result.scheme, result.netloc])
    except ValueError:
        return False

def get_all_links(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    for link in soup.find_all("a"):
        href = link.get("href")
        if href and is_valid_url(href):
            yield href
        elif href:
            href = urljoin(url, href)
            if is_valid_url(href):
                yield href
```

```python
# url = "https://www.um.edu.my/"  # Replace with the target website URL
url = websites[0]
links = list(set(get_all_links(url)))

copy_links = links.copy()


for i, link in tqdm(enumerate(copy_links), total = len(copy_links)):
#     if i > 5:
#         break
    if "www.sc.com.my" in link:
        try:
            temp_links = set(get_all_links(link))
            links.extend(temp_links)
        except Exception as e:
            print(f"Error at link {link} : {e}")

print(f"Total links: {len(links)}")
```
